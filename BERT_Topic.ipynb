#BERT Topic
using 20newsgroups data and bertopic model

```python
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))['data']
```

#Import BERTTOPIC and fit with docs

```python

model = BERTopic(language="english")
topics, probs = model.fit_transform(docs)
```

```python
model.get_topic_freq().head(5)
```

#Get Top 10 Topic Result

```python
model.get_topic(0)[:10]
```

```python
model.visualize_topics()
```

```python
model.update_topics(docs, n_gram_range=(1, 3))
```

#Search for topics that are similar to an input search_term

After having trained our model, we can use find_topics to search for topics that are similar to an input search_term.

```python
similar_topics, similarity = model.find_topics("gpu", top_n=5)
similar_topics
```
